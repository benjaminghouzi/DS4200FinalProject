```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv("Coffe_sales.csv")

# Plot
plt.figure(figsize=(8,5))
plt.hist(df["money"], bins=10, color='grey', edgecolor='black')
plt.title("Distribution of Prices")
plt.xlabel("Price")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```
